```python
import numpy as np
from dbgsom.dbgsom_ import DBGSOM
from jetstream_hugo.data import open_da
```

```python
da = open_da("ERA5", "plev", "s", "6H", (2000, 2022), "JJA", -80, 30, 20, 75)
quantizer = DBGSOM(verbose=True)
X = da.values.reshape((da.shape[0], np.prod(da.shape[1:])))
quantizer.fit(X=X)
labels_train = quantizer.classes_
labels_test = quantizer.predict(X=X)
```
